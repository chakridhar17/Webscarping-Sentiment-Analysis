```python
#Accessing the HTML content from webpage 
link1 = "https://qz.com/1572885/the-truth-behind-claims-airpods-cause-cancer/"
source1 = urllib.request.urlopen(link1).read()
soup1 = bs.BeautifulSoup(source1,'lxml')
```

```python
import re
tags = re.compile('<.*?>')
```

```python
#removing the tags of date and time
published_dateandtime1 = str(published_dateandtime1)
published_dateandtime1 = re.sub(tags,'',published_dateandtime1)
print("Date & Time -", published_dateandtime1)
```

```python
# Pull all text from the Title
title1 = soup1.find('title').text[ : ]
title1
```

```python
title1 = re.sub(tags,'',title1)
print("Title: ",title1)
```

```python
# Pull text from all instances of <p> tag within BodyText 
context1 = soup1.find_all('p')[5:18]
context1 = str(context1)
context1 = re.sub(tags,'',context1)
print(context1)
```

```python
#creating a dataframe
import pandas as pd
data1 = {'Author':Author1,'Title':title1,'Content':context1,'dateandtime':published_dateandtime1,'URL':link1}
data = pd.DataFrame([data1])
data
```

```python
#importing libraries and reading the dataset
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
%matplotlib inline
import string
import nltk
article = pd.read_csv("C:\\Users\\DELL\\Desktop\\airpodsarticle.csv")
article
```

```python
article.info()
```

```python
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)
        
    return input_txt 

#Removing the twitter handles (@user)

article['tidy_text'] = np.vectorize(remove_pattern)(article['Content'], "@[\w]*")
article.head()
```

```python
# removing special characters, numbers, punctuations
article['tidy_text'] = article['tidy_text'].str.replace("[^a-zA-Z#]", " ")
article.head()
```

```python
#removing all the words having length 3 or less
article['tidy_text'] = article['tidy_text'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))
article.head()
```

```python
#tokeniziing all the cleaned text
tokenized_article = article['tidy_text'].apply(lambda x: x.split())
tokenized_article.head()
```

```python
#stemming
from nltk.stem.porter import *
stemmer = PorterStemmer()
tokenized_article = tokenized_article.apply(lambda x: [stemmer.stem(i) for i in x]) # stemming
tokenized_article.head()
```

```python
#combining the tokenized text with tidy text
for i in range(len(tokenized_article)):
    tokenized_article[i] = ' '.join(tokenized_article[i])

article['tidy_text'] = tokenized_article
```

```python
#termfrequency
tf1 = (article['tidy_text'][1:2]).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
tf1.columns = ['words','tf']
tf1.head()
```

```python
#Inverse Document Frequency
for i,word in enumerate(tf1['words']):
  tf1.loc[i, 'idf'] = np.log(article.shape[0]/(len(article[article['tidy_text'].str.contains(word)])))

tf1.head()
```

SENTIMENT ANALYSIS

> finding the sentiment of the article using TextBlob and Vader.

```python
# import TextBlob 
from textblob import TextBlob 

content = data['Content']
```

```python
#analysing the sentiment using Vader ie using SentimentintensityAnalyser
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

def sentiment_score(sentence):
    ss = SentimentIntensityAnalyzer() 
    sentiment_dict = ss.polarity_scores(sentence)
    print("  ")
    print("Sentence was Overall Rated As", end = " ")
    if sentiment_dict['compound'] >= 0.01 : 
        print("Positive") 
        print("sentence was ", sentiment_dict['pos']*100, "% Positive") 
        
    elif sentiment_dict['compound'] < - 0.0 : 
        print("Negative") 
        print("sentence was  ", sentiment_dict['neg']*100, "% Negative") 
      
    else : 
        print("Neutral")
        print("sentence was  ", sentiment_dict['neu']*100, "% Neutral") 
```

```python
z=0
for i in content:
    a=i
    print("  ")
    print( "polarity of over all document:",z+1 )
    z=z+1
    sentiment_score(a)
```

```python
def sent_score(sentence):
    p = TextBlob(se).sentiment.polarity
    print("  ")
    print("Sentence was Overall Rated As", end = " ")
    if p >= 0.01 : 
        print("Positive") 
        print("sentence was ",p*100, "% Positive") 
        
    elif p < - 0.0 : 
        print("Negative") 
        print("sentence was  ", p*100, "% Negative") 
      
    else : 
        print("Neutral")
        print("sentence was  ", p*100, "% Neutral") 
```

```python
z=0
for i in content:
    a=i
    print("  ")
    print( "polarity of over all document:",z+1 )
    z=z+1
    sent_score(a)
```

```python
# artice polarity and finding number of sentences
from nltk.tokenize import sent_tokenize
Sentences = pd.DataFrame([])
for sentence in  content:
        sent = sent_tokenize(sentence)
        for s in sent:
            s = []
            p = TextBlob(sentence).sentiment.polarity
            if (p>=0.1):
                s = 'Positive'
            elif (p<=0.0):
                s = 'Negative'
            else:s = 'Neutral'
            p = {'Article':sentence,'Polarity': p,'Sentiment':s}
            Sentences = Sentences.append([p])
len(Sentences)
```
